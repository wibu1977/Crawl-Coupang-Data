```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
```

```python
products = []
for i in range(27):
    print('Scraping page', i+1)
    product = driver.find_elements(By.XPATH, "//div[@class='name']")
    print(f"Found {len(product)} products on page {i+1}")
    for p in product:
        products.append(p.text)
    next_button = driver.find_elements(By.XPATH, "//span[text()='다음']")
    print(f"Found {len(next_button)} next buttons on page {i+1}")
    for button in next_button:
        try:
            # Scroll the button into view
            driver.execute_script("arguments[0].scrollIntoView();", button)
            # Use an ActionChain to move to and click the button
            ActionChains(driver).move_to_element(button).click(button).perform()
            print("Clicked next button")
            sleep(2)  # Pause to allow the page to load
        except Exception as e:
            print(f"Failed to click on the button: {e}")
print(f"Scraped {len(products)} products in total")
```
